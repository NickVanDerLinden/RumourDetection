```python
# Using readline()
file1 = open('#jinek.txt', 'r')
count = 0
tweets = []

while True:
    count += 1
 
    # Get next line from file
    line = file1.readline()
 
    # if line is empty
    # end of file is reached
    if not line:
        break
    tweets.append(json.loads(line,encoding="utf-8"))
    
print(len(tweets))
texts = []
for i in range(len(tweets)):
    for j in range(len(tweets[i]["data"])):
        if tweets[i]["data"][j]["text"] not in texts:
            texts.append(tweets[i]["data"][j]["text"])
```

```python
# Using readline()
file1 = open('#vleestaks.txt', 'r')
count = 0
tweets = []

while True:
    count += 1
 
    # Get next line from file
    line = file1.readline()
 
    # if line is empty
    # end of file is reached
    if not line:
        break
    tweets.append(json.loads(line,encoding="utf-8"))
print(len(tweets))   
texts2 = []
for i in range(len(tweets)):
    for j in range(len(tweets[i]["data"])):
        if tweets[i]["data"][j]["text"] not in texts:
            texts2.append(tweets[i]["data"][j]["text"])
```

```python
# Using readline()
file1 = open('#inflatie.txt', 'r')
count = 0
tweets = []

while True:
    count += 1
 
    # Get next line from file
    line = file1.readline()
 
    # if line is empty
    # end of file is reached
    if not line:
        break
    tweets.append(json.loads(line,encoding="utf-8"))
print(len(tweets))
    
texts3 = []
for i in range(len(tweets)):
    for j in range(len(tweets[i]["data"])):
        if tweets[i]["data"][j]["text"] not in texts:
            texts3.append(tweets[i]["data"][j]["text"])
```

```python
import random

jinek_sample = random.sample(texts,750)
vleestaks_sample = random.sample(texts2,50)
inflatie_sample = random.sample(texts3, 200)
```

```python
sample = []

sample.append(jinek_sample)
sample.append(vleestaks_sample)
sample.append(inflatie_sample)

flat_list = [item for sublist in sample for item in sublist]
flat_list = [x.replace("\n", " ") for x in flat_list]
random.shuffle(flat_list)
```

```python
# open the file in the write mode
f = open('data2.csv', 'w', encoding="utf-8")

for i in flat_list[0:250]:
    f.write(str(i+"\n"))
f.close()
```
